```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Line plot
df = pd.DataFrame(np.random.randn(10,4), index = pd.date_range('1/1/2020',periods=10), columns = list('ABCD'))
```

```python
df
```

```python
df.plot()
```

```python
# Bar plot
df = pd.DataFrame(np.random.randn(10,4), columns = list('abcd'))
df.plot.bar()
```

```python
df['a'].plot.bar()
```

```python
df.plot.barh(stacked=True)
# Horizontal bar plot (stacked)
```

```python
df = pd.DataFrame({'a':np.random.randn(1000)+1, 'b':np.random.randn(1000), 'c':np.random.randn(1000)-1}, columns = list('abc'))
df.plot.hist(bins=30)
# Histogram
```

```python
plt.hist(np.random.randn(100),bins=10)
```

```python
# Box plot
df = pd.DataFrame(np.random.rand(10,5), columns = list('abcde'))
df
```

```python
df.plot.box() #helps to find IQR
```

```python
# Area plot
df.plot.area()
```

```python
# Scatter Plot
df.plot.scatter(x='a',y='b',c="red")
```

```python
# Pie Plot
df = pd.DataFrame(np.random.rand(4,1)*4)
df
```

```python
df.plot.pie(subplots=True)
```
